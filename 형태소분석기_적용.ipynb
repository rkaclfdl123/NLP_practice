# 명사 추출
    - 문자열에서 명사만 추출하려면 noun 명령을 사용한다.

<br>
<br>


```python
hannanum.nouns(c[:40])
```

```python
kkma.nouns(c[:40])
```

```python
okt.nouns(c[:40])
```

```python
komoran.nouns("\n".join([s for s in c[:40].split("\n") if s])) # 빈줄있으면 에러난다고 했는데
```

```python
komoran.nouns(c[:40]) #안남
```

#  품사 부착

```python
hannanum.pos(c[:40])
```

```python
kkma.pos(c[:40])
```

```python
# komoran은 빈줄이 있으면 에러가 남
komoran.pos("\n".join([s for s in c[:40].split("\n") if s]))
```
